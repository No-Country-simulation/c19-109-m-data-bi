```python
unique_values = {}

for col in df.columns:
  unique_values[col] = df[col].unique()

for col, value in unique_values.items():
  if len(value) <= 10: #filtro por 10 para tener una idea del dataset nomas
    print('{}: {}\n'.format(col, value))
```
